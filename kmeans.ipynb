```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv("dataset/Mall_Customers.csv")
```

```python
a
```

```python
x=a.iloc[:,3:5].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wass=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wass.append(kmeans.inertia_)
plt.plot(range(1,11),wass)
plt.xlabel("number of the cluster")
plt.ylabel("wass")
plt.title("elbow method")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++")
```

```python
pred=kmeans.fit_predict(x)
```

```python
plt.scatter(x[pred==1,0],x[pred==1,1],s=100,color="red")
plt.scatter(x[pred==2,0],x[pred==2,1],s=100,color="g")
plt.scatter(x[pred==3,0],x[pred==3,1],s=100,color="b")
plt.scatter(x[pred==4,0],x[pred==4,1],s=100,color="c")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="y",s=200)
```
