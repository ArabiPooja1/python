```python
x=[2,4,3,5,7,8,9]
y=[3,5,4,7,8,9,5]
fig=plt.figure(figsize=(6,5),facecolor='r',edgecolor='g',linewidth=22)
plt.plot(x,y)
plt.title("ALL PROPERTIES",c='g',fontsize=30)
plt.xlabel("student number",c='r',fontsize=20)
plt.ylabel("maximum 10 marks",c='r',fontsize=19)
plt.xlim(1,10)
plt.ylim(0,10)
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,10,1))
plt.legend(["osa"])
plt.show()
```

```python
x=[1,2,8,3,4,5]
y=[10,20,30,40,50,100]
fig=plt.figure(figsize=(5,5),edgecolor='r',linewidth=29,facecolor='g',)
fig.add_axes([1,1,1,1])
plt.title("Student marks",color="k",fontsize=30)
plt.xlabel("Total student",color='k',fontsize=20)
plt.ylabel("marks of the student",color='k',fontsize=20)
plt.xlim(1,10)
plt.ylim(0,10)
plt.xticks(np.arange(0,11,1))
plt.yticks(np.arange(0,110,10))
plt.legend(["osa"])
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[23,45,60,71,89,100,10]
fig=plt.figure(figsize=(18,5),facecolor='r',linewidth=22,edgecolor='g')

plt.subplot(1,4,1)
plt.plot(x,y)
plt.title("first student",size=20)

plt.subplot(1,4,2)
plt.plot(x,y)
plt.title("second student",size=19)

plt.subplot(1,4,3)
plt.title("Third student",size=20)
plt.plot(x,y)

plt.subplot(1,4,4)
plt.title("Fourth student",size=20)
plt.plot(x,y)
```

```python
x=[1,2,3,5,7,8,9,4,10]
y=[20,43,54,10,23,10,34,25,67]
fig=plt.figure(figsize=(5,3),edgecolor='r',linewidth=30,facecolor='g')

fig.add_axes([1,1,1,1])
plt.plot(x,y)
plt.title("first student")

fig.add_axes([1,2.1,1,1])
plt.plot(x,y)
plt.title("second student")

fig.add_axes([2.1,1,1,1])
plt.plot(x,y)
plt.title("third Student")

fig.add_axes([2.1,2.1,1,1])
plt.plot(x,y)
plt.title("fourth Student")
plt.show()
```

# Types of the graph

# line graph

```python
x=[1,2,3,4,5,6,7]
y=[23,43,56,7,83,45,78]
plt.plot(x,y,color='r',marker='o',markersize=10,linewidth=3,linestyle="dashed")
```

```python
x=[1,2,3,4,5,6,7]
y=[23,43,56,7,83,45,78]
plt.plot(x,y)
plt.xlabel("number of the student")
plt.ylabel("marks of the student")
```

# Bar chart

```python
x=[1,2,3,4,5,6,7]
y=[23,43,56,7,83,45,78]
plt.bar(x,y)
plt.xlabel="student",
plt.ylabel("Marks")
```

```python
x=[1,2,3,4,5,6,7]
y=[23,43,56,7,83,45,78]
plt.bar(x,y,linewidth=3,edgecolor='k',width=0.6)
x=[1,2,3,4,5,6,7]
y=[23,43,56,7,83,45,78]
plt.show()
```

# Scatter plot

```python
x=[1,2,3,4,5,6]
y=[10,20,30,40,50,60]
plt.scatter(x,y)
plt.show()
plt.show()
```

```python
x=[89,28,35,56,67]
labels=["car","mat","sun","survived","Age"]
plt.pie(x,labels=labels)
plt.show()
```

```python
x=[34,56,45,65,72,86]
labels=["car","baik","railway","van","bus","car"]
explode=[0.5,0,0,0,0,0]
colors=["brown","blue","pink","red","orange","green"]
plt.pie(x,labels=labels,explode=explode,colors=colors,autopct="%1.1f%%",shadow=True)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[23,34,56,67,81,45]
plt.plot(x,y,color='m',linestyle="dashed",linewidth=3,marker="o",markersize=10)
plt.xlabel("number of the student",color='r')
plt.ylabel("number of the marks",color='g')
plt.title("student marks",size=30,color='y')
plt.xticks(np.arange(0,11,1),color='r')
plt.yticks(np.arange(0,100,10),color='g')
plt.xlim(0,11)
plt.ylim(0,100)
plt.legend(["osa"],bbox_to_anchor=(1.2, 1.0),)
```

```python
plt.scatter(x,y,edgecolor='r',color='orange',linewidth=2,marker="o")
```

```python
x=[12,34,25,67,56,45]
explode=[0,0.2,0,0,0,0]
color=['b','g','m','r','c','y']
labels=["one","two","three","four","five","six"]
plt.pie(x,labels=labels,colors=color,explode=explode,shadow=True,autopct="%1.1f%%")
plt.show()
```
