```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv("dataset/boston_housing.csv")
```

```python
a
```

```python
plt.scatter(x="ZN",y="MEDV",data=a)
```

```python
y=a.iloc[:,-1:]
```

```python
y
```

```python
x=a.iloc[:,1:-1]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=0.20)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
```

```python
new=LinearRegression()
```

```python
new.fit(xtrain,ytrain)
```

```python
pred=new.predict(xtest)
```

```python
new.coef_
```

```python
new.intercept_
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(ytest,pred)
```

```python
mean_squared_error(ytest,pred)
```

```python
r2_score(ytest,pred)
```

```python
new.score(xtest,ytest)
```
