```python
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv("dataset/titanic.csv")
```

```python
a.head()
```

```python
a.isna().sum()
```

```python
a=a.fillna(method="ffill")
```

```python
a.drop(columns=["PassengerId","Name","Ticket","Cabin"],inplace=True)
```

```python
x=a.iloc[:,1:8]
```

```python
x
```

```python
y=a[["Survived"]]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=0.20)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
new=ColumnTransformer(transformers=[
    ("tr1",OneHotEncoder(sparse=False,drop="first"),["Sex","Embarked"])
],remainder="passthrough")
```

```python
xtrain=new.fit_transform(xtrain)
```

```python
xtest=new.fit_transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
new=LogisticRegression()
```

```python
new.fit(xtrain,ytrain)
```

```python
pred=new.predict(xtest)
```

```python
pred
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred)
```
