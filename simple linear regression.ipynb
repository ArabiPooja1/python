```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
new=ColumnTransformer(transformers=[
        ('tr1',OrdinalEncoder(categories=[["furnished","semi-furnished","unfurnished"]]),["furnishingstatus"]),
        ("tr2",OneHotEncoder(sparse=False,drop="first"),["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea"])
],remainder="passthrough")
```

```python
xtest=new.fit_transform(xtest)
```

```python
xtrain=new.fit_transform(xtrain)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
new=LinearRegression()
```

```python
new.fit(xtrain,ytrain)
```

```python
pred=new.predict(xtest)
```

```python
pred
```

```python
ytest
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ytest,pred)
```
